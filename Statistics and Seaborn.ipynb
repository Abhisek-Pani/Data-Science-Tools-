### Corelation 

```python
# So it can be done with a technique called heatmap.
```

```python
df.head()
```

```python
ff = df[['Age' , 'Fare']]
ff.head()
```

```python
ff.corr(method = 'pearson')
```

```python
# heatmap

sns.heatmap(data= ff.corr(method = 'pearson'), annot =True , cmap= 'tab20')
```

```python
sns.heatmap(data= ff.corr(method = 'pearson'), annot =True , cmap= 'crest')
```

```python
df.head()
```

```python
# apply heatmap to the entire data [python will only select numerical data columns automatically] : 

sns.heatmap(data= df.corr(method = 'pearson'), annot =True , cmap= 'tab20')
```

```python
sns.heatmap(data= ff.corr(method = 'spearman'), annot =True , cmap= 'tab20')
```

```python
sns.heatmap(data= df.corr(method = 'spearman'), annot =True , cmap= 'tab20')
```
