### Scatterplot 

```python
# used to visualize the data in 2-D format .
```

```python
plt.figure(figsize=(2,4))   # (widhth and height, this line should be at the first )
x = np.array([1,5,10,15,20])
y = np.array([10,20,30,40,50])

#using matplotlib to visualize the data :
plt.scatter(x,y , marker = '*' , color='r')
plt.title('Our Country GDP')
plt.xlabel('X-Values' , fontsize = 'xx-small')
plt.ylabel('Y-values',color = 'g' , fontsize = "large")
plt.xticks([])     # no x values are there 
plt.yticks([])     # no y values are there


plt.show()  # to see the image 
```

```python
#don't do this mistake 
x= np.arange(1,21,2)
y= np.arange(1,21,3)

print(x)
print(y)

plt.scatter(x,y)
```

### Lineplot : just join the points 

```python

plt.figure(figsize=(2,4))   
x = np.array([1,5,10,15,20])
y = np.array([10,20,30,40,50])

#using matplotlib to visualize the data :
plt.plot(x,y , marker = '+' , color='r')     # plot : means dots are joined and a line is formed 

plt.show()  # to see the image 
```

### Legend : it tries to clarify the doubts 

```python
a=[10,2,4,5]
b=[12,34,12,4]
c=[11,10,1,23]
d=[10,20,30,40]

plt.figure(figsize=(4,4))

plt.plot(a, b , label = 'A')
plt.plot(a, c, label = 'B')
plt.plot(a, d, label= 'C')

plt.legend(loc=5, fontsize='xx-small')            # loc :  used to place the legend wherever you want 
```

### subplots : we can do the visualization differently.

```python
a=[10,2,4,5]
b=[12,34,12,4]
c=[11,10,1,23]
d=[10,20,30,40]

plt.figure(figsize=(4,4))
# All in one :

plt.subplot(2,2,1)
plt.plot(a, b , label = 'A')
plt.plot(a, c, label = 'B')
plt.plot(a, d, label= 'C')
plt.title('Heroes')


#second 
plt.subplot(2,2,2)
plt.plot(a, b , label = 'A', color = 'r')
plt.title('a and b ')

#third 
plt.subplot(2,2,3)
plt.plot(a, c , label = 'B', color= 'b')
plt.title('a and c')


#fourth:

plt.subplot(2,2,4)
plt.plot(a, d , label = 'C', color = 'g')
plt.title('a and d')


plt.show()
```

## Bar_Charts :

```python
Categories = [ 'Daibetes' , 'BP' , 'Cholesterol' , 'Brain Tumuor' , 'Commoncold ']
Values = [300 , 200 , 400 , 300 , 150]

plt.figure(figsize=(8,8))
plt.bar(Categories , Values , color= 'g')
plt.xlabel('Disease' , color = 'b', size='large' )
plt.ylabel('No.of people affected per month' , color= 'r', size = 'xx-small')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4,4))
cities = [ 'USA' , 'Australia' , 'Uk']
population = [ 10000, 30000, 20000]
color = ['Green' , 'Blue' , 'Red']


plt.bar(cities, population , color = color )
plt.xlabel('Cities of most visited ')
plt.ylabel('Population')
plt.title('Country vs Population')
plt.grid(True)
plt.show()
```

### Pie chart 

```python
country = [ 'Canada' , 'India' , 'Australia' , 'England ']
Population_age18 = [100, 40 ,60 ,80]
colors = ['Green' , 'Red' , 'Blue' , 'Yellow']

plt.pie(Population_age18 , labels = country,colors = colors ,  radius = 1)
plt.show()
```

-`autopct : to put the values over the chart` `syntax as autopct = '%1.1f%%'`
-`explode : to break into pieces `

```python
country = [ 'Canada' , 'India' , 'Australia' , 'England ']
Population_age18 = [100, 40 ,60 ,80]
colors = ['Green' , 'Red' , 'Blue' , 'Yellow']
explode= [0.4,0.5,0.1,0.3]
plt.pie(Population_age18 , labels = country,colors = colors ,autopct = '%1.1f%%' ,shadow = True , radius = 1,explode = explode)
plt.show()
```

```python
country = [ 'Canada' , 'India' , 'Australia' , 'England ']
Population_age18 = [100, 40 ,60 ,80]
colors = ['Green' , 'Red' , 'Blue' , 'Yellow']
explode= [0.4,0,0,0]
plt.pie(Population_age18 , labels = country,colors = colors ,autopct = '%1.1f%%' ,shadow = True , radius = 1,explode = explode)
plt.show()
```
