### DataFrame = works like 10x

```python
a = np.arange(1,61).reshape(10,6)
a
```

```python
sol.shape
```

```python
# we can do slicing operations :

# its not numpy array but it is data frame

# we can have access to the values by loc and iloc functions 

# loc function works directly [without index] and iloc function works with index 
```

```python
# loc function :  [rowstart:rowend , colstart:colend]

sol.loc[2:4]
```

```python
sol.loc[2:4,'abra': 'la']
```

```python
sol.loc [7] [4]
```

```python
sol.loc [7 , 'la']
```

```python
sol.loc[9, 'abra':'ka']
```

```python
# iloc is also same ; same syntax [only folows n-1 rule]

sol
```

```python
sol.iloc[9::, 5:]
```

```python
sol.iloc[6:7 , 2:3]
```

```python
sol.iloc[8:9, 4:5]
```

```python
sol.iloc[6:8 , 2:4]
```

```python
# loc and iloc comparision :

# loc fun will not follow n-1 rule :

sol
```

```python
sol.loc[5:6, 'dabra':'ho']
```

```python
# iloc fun will follow n-1 rule :

sol.iloc[5:7 , 2:4]
```

### groupby 

```python
a = pd.DataFrame ({'Animal': ['Parrot','Tiger','Parrot','Tiger'] ,
                'Max speed': [200, 300, 40, 50]})
a
```

```python
a.groupby(['Animal']).mean()
```

```python
a.groupby(['Animal']).sum()
```

```python
a.groupby(['Animal']).max()
```
